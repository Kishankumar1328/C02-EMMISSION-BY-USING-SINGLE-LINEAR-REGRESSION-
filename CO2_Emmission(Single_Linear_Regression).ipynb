```python
df=d[["Engine Size(L)","Cylinders","Fuel Consumption Comb (mpg)","CO2 Emissions(g/km)"]]
df.head(13)
```

```python
plt.scatter(df["Engine Size(L)"],df["CO2 Emissions(g/km)"],color="yellowgreen")
plt.xlabel("Engine_size")
plt.ylabel("Co2_emmision")
plt.title("Emission")
```

```python
plt.scatter(df["Fuel Consumption Comb (mpg)"],df["CO2 Emissions(g/km)"],color="red")
plt.xlabel("Fuel Comsumption")
plt.ylabel("Co2_emmision")
plt.title("Emission")
```

```python
plt.scatter(df["Cylinders"],df["CO2 Emissions(g/km)"],color="violet")
plt.xlabel("Cylinders")
plt.ylabel("Co2_emmision")
plt.title("Emission")
```

```python
msk=np.random.rand(len(df))<0.8
train=df[msk]
test=df[~msk]
```

```python
plt.scatter(train["Engine Size(L)"],train["CO2 Emissions(g/km)"],color="hotpink",alpha=1)
plt.xlabel("Engine_size")
plt.ylabel("Emission")
plt.title("CO2 emmision")
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
train_x=np.asanyarray(train[["Engine Size(L)"]])
train_y=np.asanyarray(train[["CO2 Emissions(g/km)"]])
lm.fit(train_x,train_y)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
plt.scatter(train["Engine Size(L)"],train["CO2 Emissions(g/km)"],alpha=1)
plt.plot(train_x,lm.coef_[0][0]*train_x,lm.intercept_[0],"-r")
plt.title("Emission")
plt.xlabel("Engine_size")
plt.ylabel("Emission")
```

```python
from sklearn.metrics import r2_score
```

```python
test_x=np.asanyarray(test[["Engine Size(L)"]])
test_y=np.asanyarray(test[["CO2 Emissions(g/km)"]])
test_y=lm.predict(test_x)
```

```python
print("Mean Absolute Error:%.2f"%np.mean(np.absolute(test_y-test_y)))
```

```python
print("Residual Sum Of Squares(MSE):%.2f"%np.mean((test_y-test_y)**2))
```

```python
print("R2_score:%.2f"%r2_score(test_y,test_y))
```
